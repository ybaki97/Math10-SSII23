Question: Is `Series.sum()` faster than `sum(Series)`.
Answer: Yes! `sum()` is a built-in python function and has to work for a large number of data types. On the other hand, `.sum()` is specialized for pandas Series and runs very fast:
    * Series objects (like NumPy arrays) have elements all of the same data types
    * Vectorized operations (has a way of working with the entire series at once)

Here is a slightly more organized way of getting the same answer.

```python
#Or case
df[ps | ds]
```

```python
#And case
df[ps & ds]
```

## More on missing values

I want to show you a little bit of a more extreme case here. In the taxis data, pandas knew how to recognize the missing values. This will not always be the case!

```python
df2 = pd.read_csv("sampleData.csv")
df2
```

The idea here is that the question marks represent missing data, but pandas doesn't know that! This is going to cause some very big issues. Let's see some examples.

Looking at `df2`, pandas thinks that "A" and "B" represent string data...this comes from the fact that quesiton marks are in these columns.

```python
df2["A"].sum()
```

This is an example of string concatenation. It takes strings and puts them together.

```python
"15" + "16"
```

Because pandas does not recognize "A" and "C" as numeric columns, we also lose information from `describe()`

```python
df2.describe()
```

Missing values are often handled automatically by pandas, but here's how we can do it by hand, as we would need to in this case:

In this example, no rows are removed. We just replace all "?" with `nan`

```python
df3 = pd.read_csv("sampleData.csv",na_values="?")
df3
```

```python
df2
```

Notice! "A" and "C" are now recognized as columns with floating point numbers! `.sum()` and `.describe()` would work how we expect now.

## Strings in pandas revisited: Worse approach 1 (`iloc` and for-loops)

Question from earlier in the lecture:
* _How many rows in `df` have the word "Airport" in either the "pickup_zone" column or the "dropoff_zone" column. Do any of the rows have the word "Airport" in both columns?_

The solution that we wrote together earlier really is one of the best ways I know how to solve this problem. The point of this section of the notes is to get some practice with indexing and for-loops.

```python
len(df)
```

The logic of the following attemp is good, but we'll run into an error...

```python
good_inds = []

#there's almost always a better way of iterating
#than range(len(...))
for i in range(len(df)):
    if ("Airport" in df.iloc[i,10]) and ("Airport" in df.iloc[i,11]):
        good_inds.append(i)
```

Would you believe that this is an issue with missing values? Hint: What's the datatype of `np.nan`?

```python
#does this example convince you?
"Airport" in np.nan
```

The way we can get around this is by using a "try-except" block:

```python
good_inds = []

#there's almost always a better way of iterating
#than range(len(...))
for i in range(len(df)):
    try:
        if ("Airport" in df.iloc[i,10]) and ("Airport" in df.iloc[i,11]):
            good_inds.append(i)
    except TypeError:
        #go to the next step of the for-loop without breaking
        continue
```

```python
good_inds
```

```python
df.iloc[good_inds]
```

```python
#this gives an error because pandas is looking for columns with these names
df[good_inds]
```

## Strings in pandas revisited: Worse approach 3 (helper function and `map`)

We start now by writing a helper function which checks if the word "Airport" is contained in a given string.

```python
def has_airport(s):
    return "Airport" in s
```

```python
has_airport("Math 10")
```

```python
has_airport("Los Angeles International Airport")
```

Let's now define `ser1` and `ser2` to be the Series that we're looking at:

```python
ser1 = df["pickup_zone"]
ser2 = df["dropoff_zone"]
```

```python
#how does map work?
help(ser1.map)
```

Notice, we'll still get an error because of the missing values.

```python
ser1.map(has_airport)
```

Here's how we can fix it:

```python
#we see a lot of False, but that's just because we're
#looking at the first and last portions of the data
ser1.map(has_airport,na_action="ignore")
```

```python
#Putting everything together
df[ser1.map(has_airport,na_action="ignore") & ser2.map(has_airport,na_action="ignore")]
```
